```python
import pandas as pd

data = pd.read_csv(r'data\class-c beta-lactamase  cephalosporins(BLTM c3).csv', encoding='latin1', sep=';')
data
```

```python
data['Seq. '].shape[0] - data['Seq. '].isna().sum()
```

```python
seqs = [i for i in data['Seq. '][~data['Seq. '].isna()]]
seqs
```

```python
from rdkit import Chem 
from rdkit.Chem import Descriptors, rdMolDescriptors
```

```python
list_seq = [seq for seq in seqs[:-1]]
len(list_seq)
```

## Descritores moleculares

```python
def calc_seq_descriptors(seq):
    Mol = Chem.MolFromSequence(seq)
    return {'Sequence': seq,
                # Contagem de átomos e ligações
                'NumAtoms': Mol.GetNumAtoms(),
                'NumHeavyAtoms': Mol.GetNumHeavyAtoms(),
                'NumBonds': Mol.GetNumBonds(),
                'NumHeteroatoms': rdMolDescriptors.CalcNumHeteroatoms(Mol),
                'NumRotatableBonds': rdMolDescriptors.CalcNumRotatableBonds(Mol),
                'NumAromaticRings': rdMolDescriptors.CalcNumAromaticRings(Mol),
                'NumAliphaticRings': rdMolDescriptors.CalcNumAliphaticRings(Mol),
                'NumRings': Mol.GetRingInfo().NumRings(),
                
                #Propriedades derivadas da sequência
                'MolWt': Descriptors.MolWt(Mol),
                'HeavyAtomMolWt': Descriptors.HeavyAtomMolWt(Mol),
                'ExactMolWt': Descriptors.ExactMolWt(Mol),
                'ExactMolWt': Descriptors.NumValenceElectrons(Mol),
                'NumValenceElectrons': Descriptors.NumRadicalElectrons(Mol),
                'NumRadicalElectrons': Descriptors.NumRadicalElectrons(Mol),

                #Contagem de elementos e grupos funcionais
                'NumHAcceptors': Descriptors.NumHAcceptors(Mol),
                'NumHDonors': Descriptors.NumHDonors(Mol),
                'NumHeteroatoms': Descriptors.NumHeteroatoms(Mol),
                'NumSaturatedRings': Descriptors.NumSaturatedRings(Mol),
                'NumAromaticRings': Descriptors.NumAromaticRings(Mol),
                'NumAliphaticRings': Descriptors.NumAliphaticRings(Mol),
                'NumAromaticHeterocycles': Descriptors.NumAromaticHeterocycles(Mol),
                'NumAliphaticHeterocycles': Descriptors.NumAliphaticHeterocycles(Mol),

                # Propriedades físico-químicas calculadas da sequência
                'MolLogP': Descriptors.MolLogP(Mol),
                'MolMR': Descriptors.MolMR(Mol),
                'TPSA': Descriptors.TPSA(Mol),
                'FractionCSP3': Descriptors.FractionCSP3(Mol),
                'LabuteASA': Descriptors.LabuteASA(Mol)
               }
```

```python
dataset = list(map(calc_seq_descriptors, list_seq))
```

```python
dataset
```

```python
final_dataset = pd.DataFrame.from_records(dataset)
final_dataset
```
